### 2.4 Slice the intensity levels (0-255) into 6 levels.
Remember, now we need to slice the intensity levels into 6 levels in this case.

An Example of the idea of slicing is shown down below




We will do the slicing just by dvidide the intensity values of 0-255 into N (6) equal steps

You can get some inspiration of during this by using the package ```np.digitize```

An example of the idea of slicing is shown **down below**:

````python
# Generate some sample data, which could be your image data (even though it is 1D)
data = np.array([1.5, 2.5, 4.0, 5.3, 5.9, 6.0, 7.2])

# Define bin edges. The bin edges will work as those 
bin_edges = np.array([2.0, 4.0, 6.0])

# Use np.digitize to assign each data point to a bin
digitized_data = np.digitize(data, bins=bin_edges)

# Print the results
for i, value in enumerate(data):
    print(f"Data point {value} is in bin {digitized_data[i]}")
````

````python
np.unique(digitized_data)
````

Here, ```digitized_data``` indicates which bin each value in data belongs to.

* 1.5 is less than 2.0, so it's in bin 0.
* 2.5 is between 2.0 and 4.0, so it's in bin 1.
* 4.0 is exactly at the edge, so it's in bin 2.
* 5.3 is between 4.0 and 6.0, so it's in bin 2.
* 5.9 is between 4.0 and 6.0, so it's in bin 2.
* 6.0 is exactly at the edge, so it's in bin 3.
* 7.2 is greater than 6.0, so it's in bin 3.

So, the intervals of the different bins are:
* 0-2 (bin 0)
* 2-4 (bin 1)
* 4-6 (bin 2)
* 6-... (bin 3)

So, in this case when we got 6 levels of colors, we need to define a ```bin_edges``` to have a 0-5 different bins (N=6) in total
You can do that manually, or try to make a concatenated for loop using ```np.array```
